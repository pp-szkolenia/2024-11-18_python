```python
x = list(range(-10, 11))
y = [item**2 for item in x]
```

```python
fig = plt.figure()

plt.plot(x, y, label="y(x)", color="#14ff5a")
plt.plot(x, x, label="x(x)", linestyle="--")
plt.title("Tytu wykresu", fontsize=20)
plt.xlabel("Opis osi x")
plt.ylabel("Opis osi y")
plt.legend();
```

```python
CITY = "Krakow"
API_KEY = "7d0c48134ae346811fa50cf99109251f"  # wpisz swj klucz API

my_url = f"http://api.openweathermap.org/data/2.5/forecast?q={CITY}&appid={API_KEY}&units=metric"

response = requests.get(my_url)
data = response.json()

temperatures = []
timestamps = []
for item in data["list"]:
    temperatures.append(item["main"]["temp"])
    ts = item["dt_txt"]
    ts = datetime.strptime(ts, "%Y-%m-%d %H:%M:%S")
    timestamps.append(datetime.strftime(ts, "%d.%m %H:%M"))
    # print(item["dt_txt"], item["main"]["temp"], sep=' | ')
```

```python
plt.plot(timestamps, temperatures, color="green")
plt.title("Prognoza temperatury")

plt.xticks(list(range(0, len(timestamps), 5)), rotation=45);
```
